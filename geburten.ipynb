```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.patches as mpatches
```

```python
df = pd.read_csv("geburten_bereinigt.csv", delimiter=",")
df.head()
```

```python
df.columns = df.columns.str.lower()
```

```python
df.geburten = df.geburten.replace("-",0)
df.geburten = df.geburten.replace(".",0)
df.geburten.unique()
```

```python

df.geburten = df.geburten.astype("int64")
df.dtypes
```

```python

plt.figure(figsize=(20,10))
sns.lineplot(data=df, x="jahr", y="geburten")
plt.xlabel("Jahre")
plt.ylabel("Geburten pro Jahr")

```

```python
plt.figure(figsize=(20,10))
sns.histplot(data=df, x="geburten")
```

```python
plt.figure(figsize=(25,8))
box = sns.boxenplot(data=df, x="stadtteilname", y="geburten")
plt.xlabel("Stadtteile")
plt.ylabel("Geburten pro Stadtteil")
box.set_xticklabels(box.get_xticklabels(),
rotation=45,
horizontalalignment='right',
fontsize='x-large')
```

```python
df.jahr.unique()
df15 = df[(df.jahr == 2015)]
df16 = df[(df.jahr == 2016)]
df17 = df[(df.jahr == 2017)]
df18 = df[(df.jahr == 2018)]
df19 = df[(df.jahr == 2019)]
df20 = df[(df.jahr == 2020)]
df21 = df[(df.jahr == 2021)]
df22 = df[(df.jahr == 2022)]
```

```python
sns.set_theme(style='dark')
plt.figure(figsize=(25,12))
box = sns.barplot(data=df15, x="stadtteilname", y="geburten", color="green",linewidth=2)
sns.barplot(data=df16, x="stadtteilname", y="geburten", color="lime",linewidth=2)
sns.barplot(data=df17, x="stadtteilname", y="geburten", color="seagreen",linewidth=2)
sns.barplot(data=df18, x="stadtteilname", y="geburten", color="mediumseagreen",linewidth=2)
sns.barplot(data=df19, x="stadtteilname", y="geburten", color="springgreen",linewidth=2)
sns.barplot(data=df20, x="stadtteilname", y="geburten", color="mintcream",linewidth=2)
sns.barplot(data=df21, x="stadtteilname", y="geburten", color="mediumspringgreen",linewidth=2)
sns.barplot(data=df22, x="stadtteilname", y="geburten", color="mediumaquamarine",linewidth=2)
df15l = mpatches.Patch(color="green", label="2015")
df16l = mpatches.Patch(color="lime", label="2016")
df17l = mpatches.Patch(color="seagreen", label="2017")
df18l = mpatches.Patch(color="mediumseagreen", label="2018")
df19l = mpatches.Patch(color="springgreen", label="2019")
df20l = mpatches.Patch(color="mintcream", label="2020")
df21l = mpatches.Patch(color="mediumspringgreen", label="2021")
df22l = mpatches.Patch(color="mediumaquamarine", label="2022")

plt.xlabel("Stadtteile")
plt.ylabel("Geburten pro Stadtteil")
plt.legend(handles=[df22l,df21l,df20l,df19l,df18l,df17l,df16l,df15l])
box.set_xticklabels(box.get_xticklabels(),
rotation=45,
horizontalalignment='right',
fontsize='x-large')
```

```python
plt.figure(figsize=(20,10))
plt.subplot(2,4,1)
sns.histplot(data=df15, x="geburten", color="green")
plt.subplot(2,4,2)
sns.histplot(data=df16, x="geburten", color="lime")
plt.subplot(2,4,3)
sns.histplot(data=df17, x="geburten", color="seagreen")
plt.subplot(2,4,4)
sns.histplot(data=df18, x="geburten", color="mediumseagreen")
plt.subplot(2,4,5)
sns.histplot(data=df19, x="geburten", color="springgreen")
plt.subplot(2,4,6)
sns.histplot(data=df20, x="geburten", color="yellowgreen")
plt.subplot(2,4,7)
sns.histplot(data=df21, x="geburten", color="mediumspringgreen")
plt.subplot(2,4,8)
sns.histplot(data=df22, x="geburten", color="mediumaquamarine")
```
